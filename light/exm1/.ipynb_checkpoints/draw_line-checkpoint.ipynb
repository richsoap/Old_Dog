```python
import matplotlib.pyplot as plt
import numpy as np
def lineplot(x_data, y_data, x_label="", y_label="", title="", ax=""):
    if(ax == ""):
        _, ax = plt.subplots()
    ax.plot(x_data, y_data, lw = 2, color = '#539caf', alpha = 1)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    return ax
```

```python
from sklearn import linear_model
def line_regr(x_data, y_data, x_new):
    regr = linear_model.LinearRegression()
    regr.fit(x_data, y_data)
    return regr.predict(x_new)
```

```python
x = np.linspace(-1,1,50)
y = 2*x+1
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import interpolate
fig = plt.gcf()
fig.set_size_inches(10,10)
x = [20,24,27,28,29,30,31,32,34,38,42,44,46,50,54,57,61]
x_ = np.array(x).reshape(-1,1)
y = [0.00101,0.00101,0.00101,0.00222,0.00915,0.0144,0.0181,0.0238,0.0274,0.0483,0.0626,0.0658,0.0988,0.114,0.118,0.132,0.137,]
y_ = np.array(y).reshape(-1,1)

new_x = np.linspace(20,60,50).reshape(-1,1)

regr = linear_model.LinearRegression()
regr.fit(x_, y_)
a, b = regr.coef_, regr.intercept_
c = -b/a
print(c)
new_y = regr.predict(new_x)
smooth = interpolate.interp1d(x,y,kind='cubic')
smooth_y = smooth(new_x)
plt.plot(new_x, smooth_y, lw = 2, color = '#ff0000', alpha = 1)
plt.xlabel("I/mA")
plt.ylabel("P/mW")
plt.plot(new_x,new_y)
plt.show()
fig.savefig('exm1_1.eps', dpi=100)
```
